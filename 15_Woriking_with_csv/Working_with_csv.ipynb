## 3. Opening the csv file with url

```python

import pandas as pd
import requests
from io import StringIO

url = "https://example.com/data.csv"  # Replace with actual CSV URL

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "text/csv",
}

req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)
print(df.head())
```

## 10. Encoding parameter

```python
# By default, pandas.read_csv() assumes the file is encoded in UTF-8, which is the most common encoding for CSV files. However, if your file is in a different encoding, you may need to specify it explicitly.

import chardet

with open("Titanic-Dataset.csv", "rb") as file:
    result = chardet.detect(file.read(10000))  # Read a portion to detect encoding

print(result['encoding'])  # Prints detected encoding, e.g., 'utf-8'
```

```python
pd.read_csv("Titanic-Dataset.csv", encoding = "latin-1")
```

## 14. Converters

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
    
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv("matches.csv", converters = {'team1':rename} )
```
