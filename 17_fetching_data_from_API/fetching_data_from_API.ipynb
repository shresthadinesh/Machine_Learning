```python
for i in range(1,499):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=bec0154cc7c620f32ca7f0faa37ff529&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json().get('results', []), columns=['id', 'title', 'release_date', 'popularity', 'vote_average', 'vote_count'])
    df = pd.concat([df, temp_df], ignore_index=True)
    
```
