```python
Position = []
Company_name = []
Address = []

Position = [i.find('h1', class_='text-primary').text.strip() for i in company if i.find('h1', class_='text-primary')]
Company_name = [i.find('h3', class_ = 'h6').text.strip() for i in company if i.find('h3', class_ = 'h6')]
Address = [i.find('div', class_ = 'media-body').text.strip() for i in company if i.find('div', class_ = 'media-body')]

d = {'Position': Position, 'Company_name' : Company_name, 'Address' : Address}
df = pd.DataFrame(d)
```

```python
df
# This is the data of page=1
```

```python
final = pd.DataFrame()

for j in range(1,12):

    url = "https://merojob.com/category/it-telecommunication/?page={}".format(j)

    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage, "lxml")

    company = soup.find_all('div', class_="card-body")

    Position = []
    Company_name = []
    Address = []

    Position = [i.find('h1', class_='text-primary').text.strip() for i in company if i.find('h1', class_='text-primary')]
    Company_name = [i.find('h3', class_ = 'h6').text.strip() for i in company if i.find('h3', class_ = 'h6')]
    Address = [i.find('div', class_ = 'media-body').text.strip() for i in company if i.find('div', class_ = 'media-body')]

    d = {'Position': Position, 'Company_name' : Company_name, 'Address' : Address}
    df = pd.DataFrame(d)
    
    final = pd.concat([final, df], ignore_index=True)
            
```
